# 2.Descriptive statistics


```python
#descriptive statistics
df.describe()
```

```python
#mean, std, min, max, range , median, mode for salary
print(df['TotalPay'].describe())
print("the range for salary : ", df.TotalPay.min(), df.TotalPay.max())
print("the median for salary : ", df.TotalPay.median())
print("the mode for salary : ", df.TotalPay.mode())
```

# 3.Data cleaning


```python
#dropped the columns Notes and Status by using drop function as they contain null values
df = df.drop(['Notes' , 'Status'], axis=1) #no 1 for columns
df.head()
```

```python
#fill in missing values with mean of column
df['Benefits'].fillna(df['Benefits'].mean(), inplace = True)
df.head()
```

```python
df['BasePay'].fillna(df['BasePay'].mean(), inplace = True)
df['OvertimePay'].fillna(df['OvertimePay'].mean(), inplace = True)
df['OtherPay'].fillna(df['OtherPay'].mean(), inplace = True)
```

```python
df.isna().sum()
```

```python
#remove negative values
df=df[(df["BasePay"] > 0) &(df["OvertimePay"] > 0) & (df["TotalPay"] > 0) & (df["OtherPay"] > 0)]
df.describe()
```
